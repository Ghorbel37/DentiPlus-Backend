### Generate dataset bigger

```python
import random
import pandas as pd
import string

# --- Define templates and word lists for each class ---

# Class 0: Symptom descriptions
symptom_templates_en = [
    "I have {} in my {}.",
    "Since yesterday, I've been experiencing {} in my {}.",
    "Lately, I feel {} whenever I {}.",
    "My {} hurts and I feel {}.",
    "There is a {} pain in my {}.",
    "I'm suffering from {} in my {}.",
    "I feel {} in my {} every time I {}.",
    "The {} in my {} is {} today."
]

symptom_adjectives_en = ["a sharp", "a dull", "an intermittent", "a constant", "a burning", "a throbbing", "a mild", "a severe"]
symptom_bodyparts_en = ["head", "stomach", "chest", "back", "leg", "arm", "ear", "throat", "abdomen", "knee", "shoulder", "wrist"]
symptom_actions_en = ["move", "walk", "stand", "run", "cough", "eat", "breathe", "sleep", "jump", "bend"]

# Class 0: French symptoms
symptom_templates_fr = [
    "J'ai {} dans mon {}.",
    "Depuis hier, je ressens {} dans mon {}.",
    "RÃ©cemment, j'ai {} quand je {}.",
    "Mon {} me fait mal et je ressens {}.",
    "Il y a une douleur {} dans mon {}.",
    "Je souffre de {} dans mon {}.",
    "Je sens {} dans mon {} chaque fois que je {}.",
    "La {} de mon {} est {} aujourd'hui."
]

symptom_adjectives_fr = ["une douleur aiguÃ«", "une douleur sourde", "une douleur intermittente", "une douleur constante", "une sensation de brÃ»lure", "une douleur lancinante", "une douleur modÃ©rÃ©e", "une douleur sÃ©vÃ¨re"]
symptom_bodyparts_fr = ["tÃªte", "estomac", "poitrine", "dos", "jambe", "bras", "oreille", "gorge", "abdomen", "genou", "Ã©paule", "poignet"]
symptom_actions_fr = ["marcher", "se lever", "s'asseoir", "manger", "respirer", "dormir", "courir", "sauter", "se pencher"]

# Class 1: Diagnosis request
conclusion_templates = [
    "Can you tell me what is wrong?",
    "I think that's all, can you provide a diagnosis?",
    "Please conclude and tell me your diagnosis.",
    "Is that enough information for a diagnosis?",
    "What do you think I have?",
    "I believe I've described all my symptoms, what's the diagnosis?",
    "Could you now give me the result?",
    "I have no more symptoms; please diagnose me.",
    "Do you have a diagnosis for me?",
    "Can you sum up what might be the issue?"
]

conclusion_templates_fr = [
    "Peux-tu me dire ce qui ne va pas ?",
    "Je pense que c'est tout, peux-tu donner un diagnostic ?",
    "Merci de conclure et de me donner un diagnostic.",
    "Est-ce assez pour Ã©tablir un diagnostic ?",
    "Que penses-tu que j'ai ?",
    "J'ai dÃ©crit tous mes symptÃ´mes, quel est le diagnostic ?",
    "Peux-tu me donner le rÃ©sultat ?",
    "Je n'ai plus de symptÃ´mes, fais ton diagnostic.",
    "As-tu un diagnostic pour moi ?",
    "Peux-tu rÃ©sumer ce qui pourrait poser problÃ¨me ?"
]

# Class 2: Polite / neutral
polite_templates = [
    "Thank you, doctor.",
    "Thanks for your help!",
    "Goodbye.",
    "See you later.",
    "Nice talking to you.",
    "I appreciate your assistance.",
    "Hello, how are you?",
    "Good morning.",
    "I really appreciate your time.",
    "Thanks a lot!",
    "Take care, doctor."
]

polite_templates_fr = [
    "Merci, docteur.",
    "Merci pour votre aide!",
    "Au revoir.",
    "Ã€ bientÃ´t.",
    "Content de vous parler.",
    "J'apprÃ©cie votre assistance.",
    "Bonjour, comment Ã§a va?",
    "Bon matin.",
    "Je vous remercie pour votre temps.",
    "Merci beaucoup!",
    "Prenez soin de vous, docteur."
]

# --- Template helper ---
def count_placeholders(template):
    return sum(1 for _ in string.Formatter().parse(template) if _[1] is not None)

# --- Generate examples ---
def generate_symptom_example():
    if random.random() < 0.5:  # English
        template = random.choice(symptom_templates_en)
        num_slots = count_placeholders(template)
        if num_slots == 2:
            return template.format(random.choice(symptom_adjectives_en), random.choice(symptom_bodyparts_en))
        elif num_slots == 3:
            return template.format(random.choice(symptom_adjectives_en), random.choice(symptom_bodyparts_en), random.choice(symptom_actions_en))
        else:
            return "Invalid template"
    else:  # French
        template = random.choice(symptom_templates_fr)
        num_slots = count_placeholders(template)
        if num_slots == 2:
            return template.format(random.choice(symptom_adjectives_fr), random.choice(symptom_bodyparts_fr))
        elif num_slots == 3:
            return template.format(random.choice(symptom_adjectives_fr), random.choice(symptom_bodyparts_fr), random.choice(symptom_actions_fr))
        else:
            return "ModÃ¨le invalide"

def generate_conclusion_example():
    return random.choice(conclusion_templates + conclusion_templates_fr)

def generate_polite_example():
    return random.choice(polite_templates + polite_templates_fr)

# --- Build dataset ---
num_examples_per_class = 1500 * 3  # Tripled

texts = []
labels = []

for _ in range(num_examples_per_class):
    texts.append(generate_symptom_example())
    labels.append(0)

for _ in range(num_examples_per_class):
    texts.append(generate_conclusion_example())
    labels.append(1)

for _ in range(num_examples_per_class):
    texts.append(generate_polite_example())
    labels.append(2)

# Shuffle dataset
combined = list(zip(texts, labels))
random.shuffle(combined)
texts, labels = zip(*combined)

# Create DataFrame
df = pd.DataFrame({"text": texts, "label": labels})
print("Dataset size:", df.shape)

# Remove duplicates
df = df.drop_duplicates()
print("After removing duplicates:", df.shape)

# Save to CSV
csv_filename = "/content/drive/MyDrive/dataset/custom_medical_dataset.csv"
df.to_csv(csv_filename, index=False, encoding='utf-8-sig')
print(f"âœ… Dataset saved as: {csv_filename}")
print(df.sample(5))

```

### Train model on generated dataset and save it

```python
import pandas as pd
from datasets import Dataset

# Remplace le chemin par le tien
file_path = "/content/drive/MyDrive/dataset/custom_medical_dataset.csv"

# Lire le CSV (on prÃ©cise l'encodage pour corriger les caractÃ¨res spÃ©ciaux)
df = pd.read_csv(file_path, encoding="utf-8")

# VÃ©rification rapide
print(df.head())

# CrÃ©ation du Dataset Hugging Face
full_dataset = Dataset.from_pandas(df)
```

```python
# Split alÃ©atoire en train/test (80/20)
dataset_split = full_dataset.shuffle(seed=42).train_test_split(test_size=0.2)
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification, TrainingArguments, Trainer
import torch

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

model_checkpoint = "distilbert-base-uncased"
tokenizer = AutoTokenizer.from_pretrained(model_checkpoint)

def tokenize(batch):
    return tokenizer(batch["text"], padding=True, truncation=True)

tokenized_datasets = dataset_split.map(tokenize, batched=True)

nlp_model = AutoModelForSequenceClassification.from_pretrained(model_checkpoint, num_labels=3)
nlp_model = nlp_model.to(device)

training_args = TrainingArguments(
    output_dir="checkpoints",
    eval_strategy="epoch",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    num_train_epochs=3,
    logging_steps=10,
    save_strategy="no"
)

trainer = Trainer(
    model=nlp_model,
    args=training_args,
    train_dataset=tokenized_datasets["train"],
    eval_dataset=tokenized_datasets["test"],
    tokenizer=tokenizer
)

trainer.train()

# After trainer.train()
nlp_model.save_pretrained("/content/drive/MyDrive/dataset/symptoms_classifier")
tokenizer.save_pretrained("/content/drive/MyDrive/dataset/symptoms_classifier")

def check_end_of_conversation_nlp(text: str) -> bool:
    nlp_model.eval()  # Set model to evaluation mode
    inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True)
    inputs = {k: v.to(device) for k, v in inputs.items()}
    with torch.no_grad():
        logits = nlp_model(**inputs).logits
    predicted_class = torch.argmax(logits, dim=1).item()
    return predicted_class


test_text = "Hi, my name is mahdi"
print(check_end_of_conversation_nlp(test_text))
if check_end_of_conversation_nlp(test_text)==1:
    print("âœ… Conversation is ready to end and diagnose.")
else:
    print("ðŸ• Still gathering symptoms.")
```
